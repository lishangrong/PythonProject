```python
from matplotlib.axes import Axes
import matplotlib.pyplot as plt

# 展示中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 创建子图
# figure：画布对象，axes：子图数组
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6), dpi=100)
# 图1：柱状图（世界石油储备）
countries = ['中国', '美国', '日本', '印度', '法国','伊拉克', '沙特']
values = [35, 23, 15, 18, 12, 48, 37]
axes1:Axes = axes[0]
axes1.bar(countries, values, color='#FFA500', width=0.5) #柱状图
axes1.set_title('世界石油储备', fontsize=18) #标题
axes1.set_xlabel('国家', fontsize=14, loc='left') #X轴标签
axes1.set_ylabel('石油储备(亿吨)', fontsize=14, loc='top') #y轴标签
axes1.grid(linestyle='--', alpha=0.3) #网格线

# 图2：饼状图（世界人口）
countries2 = ["印度", "中国", "美国","印尼", "巴基斯坦","尼日利亚","巴西", "俄罗斯", "其他"]
values = [14.51, 14.09, 3.4, 2.83, 2.51, 2.33, 2.12, 1.44, 20]

axes2: Axes = axes[1]
axes2.pie(values, labels=countries2, autopct="%1.1f%%") #饼状图
axes2.set_title("世界人口", fontsize=18) #标题
#图例 ncol: 表示图例每行显示个数,  bbox_to_anchor: 表示图例位置(0.5 控制水平方位， -0.1 控制垂直方位)
axes2.legend(loc="lower center", ncol=5, bbox_to_anchor=(0.5, -0.1))

# 保存图片
plt.savefig('data/01.png')

plt.show() #展示图表
```
