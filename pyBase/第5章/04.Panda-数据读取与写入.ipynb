```python
df3 = pd.read_csv("data/sales.csv")
#配置项，展示所有的数据
pd.set_option("display.max_rows", None)
# 数据过滤
# 1.获取销售数量>=10的订单数据
df3[df3["销售数量"] >= 10]

# 2.获取产品类别分别为 食品或图书的订单数据
df3[df3["产品类别"].isin(["食品", "图书"])]

# 3.获取单价在 100-200之间的订单数据  between(默认 包含边界)
# 设置边界：inclusive="both"| "neither"| "left" |"right"
df3[df3["单价"].between(79, 199, inclusive="neither")]

# 4. 获取 销售数量 >=8, 并且 单价 >=100 的订单数据  ------>多条件(并且： &， 或： |)
df3[(df3["销售数量"] >=8) & (df3["单价"] >=100)]

# 5. 获取 产品类别为 服装/食品，支付方式为 支付宝/微信支付 的订单数据
df3[(df3["产品类别"].isin(["服装", "食品"])) & (df3["支付方式"].isin(["支付宝", "微信支付"]))]

```
